**Create Vocabulary**

```python
import nltk
import string
import numpy as np
from nltk.corpus import stopwords
nltk.download('stopwords')
print(stopwords.words('english'))
```

```python
def remove_punctuations(text: str) -> str:
  return text.translate(str.maketrans('', '', string.punctuation))
def remove_stopwords(text: str) -> str:
  for word in stopwords.words('english'):
    new_text = text
    if word in new_text:
        new_text = new_text.replace(word, '')
  return new_text
def normalize_text(text: str) -> str:
  normalized_text = text.lower()
  normalized_text = remove_punctuations(normalized_text)
  normalized_text = remove_stopwords(normalized_text)
  return normalized_text
```

```python
import os 
file_name = os.listdir('/content/Data')
file_name.remove('.git')
file_name.remove('README.md')
file_name.remove('.gitignore')
file_name
```

```python
doc_lists = []
vocab = []
import io
root_folder = '/content/Data'
for folder in file_name:
  folder_path = os.path.join(root_folder, folder)
  for corpus in os.scandir(folder_path):
    file_path = os.path.join(folder_path, corpus)
    print(corpus)
    with io.open(file_path, 'rb') as f:
      lines = f.readlines()
      title = lines[0].strip()
      article = b' '.join(lines[1:]).strip()
      article = article.decode('utf-8')
      article = normalize_text(article)
      if (title, article) not in doc_lists:
        doc_lists.append((title, article))
      tokens = article.split(' ')
      for token in tokens:
        if token not in vocab:
          vocab.append(token)
```

```python
len(vocab)
```

```python
from gensim.models import Word2Vec
```

```python
sentences = []
for corpus in doc_lists:
  rows = corpus[1].split('/n')
  for sentence in rows:
    words = sentence.split()
    for i in range(0,len(words)-5, 5):
      sentences.append(words[i:i+5])
```

```python
sentences
```

```python
model = Word2Vec(sentences, min_count = 1, window = 10, sg = 1, workers = 8)
```

```python
model.train(sentences, total_examples = model.corpus_count, epochs = 10)
```

```python
print('embedding vector shape: ', model.wv['bark'].shape)
```

```python
model.most_similar('alba')
```

```python
def cosine_similarity(vector_a, vector_b):
    np_vector_a = np.array(vector_a)
    np_vector_b = np.array(vector_b)
    return np.dot(np_vector_a, np_vector_b)/(np.linalg.norm(np_vector_a) * np.linalg.norm(np_vector_b))
query_embedding = model.wv['basella'] + model.wv['alba']
```

```python
results = []
for corpus in doc_lists:
  temp_list = []
  vector = np.zeros(100)
  count = 0
  for word in corpus[1].split():
    try:
      vector += model.wv[word]
      count += 1
    except:
      print("error")
  vector = vector / count
  temp_list.append(corpus[0])
  temp_list.append(corpus[1])
  temp_list.append(cosine_similarity(query_embedding, vector))
  results.append(temp_list)
```

```python
results.sort(reverse = True, key = lambda x:x[2])
final_results = results[:10]
```

```python
for corpus in final_results:
  print(corpus[0])
```
