```python
import nltk
import string
import numpy as np
from nltk.corpus import stopwords
nltk.download('stopwords')
print(stopwords.words('english'))
```

```python
def remove_punctuations(text: str) -> str:
  return text.translate(str.maketrans('', '', string.punctuation))
def remove_stopwords(text: str) -> str:
  for word in stopwords.words('english'):
    new_text = text
    if word in new_text:
        new_text = new_text.replace(word, '')
  return new_text
def normalize_text(text: str) -> str:
  normalized_text = text.lower()
  normalized_text = remove_punctuations(normalized_text)
  normalized_text = remove_stopwords(normalized_text)
  return normalized_text
```

```python
def vectorize(article, vocab):
  vector = []
  res = []
  text = normalize_text(article)
  for word in vocab:
    res.append(text.count(word))
  return res
```

```python
def create_docterm_matrix(doc_lists, vocab):
  docterm_matrix = {}
  for (title,article) in doc_lists:
    vector = vectorize(article, vocab)
    docterm_matrix[title, article] = vector
  return docterm_matrix
```

```python
def calculate_cosine_similarity(vector_a, vector_b):
  np_vector_a = np.array(vector_a)
  np_vector_b = np.array(vector_b)
  return np.dot(np_vector_a, np_vector_b)/(np.linalg.norm(np_vector_a) * np.linalg.norm(np_vector_b))
```

```python
def ranking(docterm_matrix, vector_query):
  ranking = []
  for doc_info, vector in docterm_matrix.items():
    cosine_similarity = calculate_cosine_similarity(vector_query, vector)
    ranking.append((cosine_similarity,doc_info[0]))    
  return sorted(ranking, reverse = True)
```

```python
import os 
file_name = os.listdir('/content/Data')
file_name.remove('.git')
file_name.remove('README.md')
file_name.remove('.gitignore')
file_name
```

```python
doc_lists = []
vocab = []
import io
root_folder = '/content/Data'
for folder in file_name:
  folder_path = os.path.join(root_folder, folder)
  for corpus in os.scandir(folder_path):
    file_path = os.path.join(folder_path, corpus)
    print(corpus)
    with io.open(file_path, 'rb') as f:
      lines = f.readlines()
      title = lines[0].strip()
      article = b' '.join(lines[1:]).strip()
      article = article.decode('utf-8')
      article = normalize_text(article)
      if (title, article) not in doc_lists:
        doc_lists.append((title, article))
      tokens = article.split(' ')
      for token in tokens:
        if token not in vocab:
          vocab.append(token)
```

```python
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer('distilbert-base-nli-stsb-mean-tokens')
```

```python
docterm_matrix = []
for i in range(len(doc_lists)):
  temp_list = []
  temp_list.append(doc_lists[i][0])
  temp_list.append(doc_lists[i][1])
  temp_list.append(model.encode(doc_lists[i][1]))
  docterm_matrix.append(temp_list)
```

```python
query = "Alpinia Galanga"
query_embedding = model.encode(query)
```

```python
def euclidean_distance(vector_a, vector_b):
  return np.linalg.norm(vector_a - vector_b)
```

```python
results = []
for corpus in docterm_matrix:
  cosine_value = calculate_cosine_similarity(corpus[2], query_embedding)
  results.append((cosine_value, corpus))
```

```python
results = sorted(results, reverse = True)
```

```python
final_results = []
for corpus in results:
  for word in query.lower().split():
    try:
      title = corpus[1][0].decode("utf-8")
    except:
      print("Can't Decode")
    if word in title.lower():
      final_results.append(corpus)
```
